```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/wine-variety-data/OSX_DS_assignment.csv')
df
```

```python
df['variety'].value_counts()
```

```python
sns.barplot(x=df['variety'], y=df['price'])
```

```python
df_imp_features = df.drop(columns=['user_name', 'review_title', 'review_description', 'points', 'price', 'region_2'], axis=1)
df_imp_features
```

```python
df_imp_features.isna().sum()
```

```python
df_imp_features = df_imp_features.dropna()
```

```python
X = df_imp_features.drop(columns=['variety'], axis=1)
X.head(5)
```

```python
y = df_imp_features['variety']
```

```python
X.isna().sum()
```

```python
y.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = OneHotEncoder(handle_unknown = 'ignore')
le = LabelEncoder()
```

```python
encoder.fit(X_train)
le.fit(y_train)
```

```python
X_train_encoded = encoder.transform(X_train)
y_train_encoded = le.transform(y_train)

X_test_encoded = encoder.transform(X_test)
y_test_encoded = le.transform(y_test)
```

```python
X_train_encoded = X_train_encoded.toarray()
X_test_encoded = X_test_encoded.toarray()
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_forest = RandomForestClassifier()
```

```python
model_forest.fit(X_train_encoded, y_train_encoded)
```

```python
y_pred_train = model_forest.predict(X_train_encoded)
train_accuracy = accuracy_score(y_train_encoded, y_pred_train)
print(f"Training Accuracy: {train_accuracy}")
```

```python
y_pred_test = model_forest.predict(X_test_encoded)
test_accuracy = accuracy_score(y_test_encoded, y_pred_test)
print(f"Test Accuracy: {test_accuracy}")
```
